```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sbn
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score, recall_score, precision_score, roc_curve, roc_auc_score

Data = pd.read_csv("/kaggle/input/wine-dataset/wine_dataset.csv")
pd.set_option("display.max_columns", 13)

Data["style"] = np.where(Data["style"] == "white", 1, 0)

x = Data.drop(["style"], axis = 1)
y = Data["style"]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.33, random_state = 45)

model = GaussianNB()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Recall: {recall_score(y_test, y_pred)}")
print(f"Precision: {precision_score(y_test, y_pred)}")
```
